```python
plt.figure(figsize=(15,6))
explode = [0.2,0.01]
colors = sns.color_palette('bright')
plt.pie(data['Potability'].value_counts(), labels=['Potable', 'Not Potable'],
colors = colors, autopct = '%0.0f%%', explode = explode, shadow = 'True',
startangle = 90)
plt.show()
```

```python
not_potable = data.query("Potability == 0")
potable = data.query("Potability == 1")
plt.figure(figsize = (15, 15))
for ax, col in enumerate(data.columns[:9]):
    plt.subplot(3, 3, ax + 1)
    plt.title(col)
    sns.kdeplot(x = not_potable[col], label = "Not Potable")
    sns.kdeplot(x = potable[col], label = "Potable")
    plt.legend()
plt.tight_layout()
```

```python
x = data.drop("Potability", axis=1).values
y = data["Potability"].values
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state=0)
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
print("Training Accuracy:", classifier.score(x_train,y_train))
print("Test Accuracy:", classifier.score(x_test,y_test))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```
